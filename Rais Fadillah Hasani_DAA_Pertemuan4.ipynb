```python
def swap(var1, var2):
  """
  Fungsi untuk menukar nilai dua variabel.

  Args:
    var1: Variabel pertama.
    var2: Variabel kedua.

  Returns:
    Tuple berisi nilai var1 dan var2 yang telah ditukar.
  """
  var3 = var1
  var1 = var2
  var2 = var3
  return var1, var2
```

```python

def bubble_sort(list_):
  n = len(list_)
  for i in range(n-1):
    for j in range(n-i-1):
      if list_[j] > list_[j+1]:
        list_[j], list_[j+1] = swap(list_[j], list_[j+1])
  return list_

numbers = [100, 20, 60, 90, 40, 30, 10]
sorted_numbers = bubble_sort(numbers)
print("Deret bilangan yang telah diurutkan:", sorted_numbers)
```

```python
def insertion_sort(list_):
  """
  Fungsi untuk mengurutkan deret bilangan menggunakan metode insertion sort.

  Args:
    list_: Daftar bilangan yang akan diurutkan.

  Returns:
    Daftar bilangan yang telah diurutkan.
  """
  for i in range(1, len(list_)):
    key = list_[i]
    j = i - 1
    while j >= 0 and key < list_[j]:
      list_[j + 1] = list_[j]
      j -= 1
    list_[j + 1] = key
  return list_

numbers = [89, 12, 57, 16, 25, 11, 75]
sorted_numbers = insertion_sort(numbers)
print("Deret bilangan yang telah diurutkan:", sorted_numbers)
```

```python
def selection_sort(list_):
  """
  Fungsi untuk mengurutkan deret bilangan menggunakan metode selection sort.

  Args:
    list_: Daftar bilangan yang akan diurutkan.

  Returns:
    Daftar bilangan yang telah diurutkan.
  """
  n = len(list_)
  for i in range(n):
    min_idx = i
    for j in range(i + 1, n):
      if list_[min_idx] > list_[j]:
        min_idx = j
    list_[i], list_[min_idx] = list_[min_idx], list_[i]
  return list_


numbers = [89, 12, 57, 16, 25]
sorted_numbers = selection_sort(numbers)
print("Deret bilangan yang telah diurutkan:", sorted_numbers)
```

```python
def linear_search(deret, target):
  """
  Fungsi untuk mencari elemen dalam deret menggunakan metode linear search.

  Args:
    deret: Deret elemen yang akan dicari.
    target: Elemen yang akan dicari.

  Returns:
    Indeks elemen yang dicari jika ditemukan, jika tidak ditemukan mengembalikan -1.
  """
  for i in range(len(deret)):
    if deret[i] == target:
      return i
  return -1

deret_huruf = ['y', 'u', 'i', 'w', 'o', 'a', 'q', 'u', 'j', 'p']
target = 'a'

index = linear_search(deret_huruf, target)

if index != -1:
  print(f"Elemen '{target}' ditemukan pada indeks {index}.")
else:
  print(f"Elemen '{target}' tidak ditemukan dalam deret.")
```

```python


def binary_search(list_, target):
  """
  Fungsi untuk mencari elemen dalam daftar yang telah diurutkan menggunakan metode binary search.

  Args:
    list_: Daftar yang telah diurutkan yang akan dicari.
    target: Elemen yang akan dicari.

  Returns:
    Indeks elemen jika ditemukan, -1 jika tidak ditemukan.
  """
  low = 0
  high = len(list_) - 1

  while low <= high:
    mid = (low + high) // 2
    if list_[mid] == target:
      return mid
    elif list_[mid] < target:
      low = mid + 1
    else:
      high = mid - 1

  return -1


deret_huruf = ['a', 'i', 'j', 'o', 'p', 'q', 'u', 'u', 'w', 'y']  # Pastikan deret huruf terurut
target = 'a'

indeks = binary_search(deret_huruf, target)

if indeks != -1:
  print(f"Huruf '{target}' ditemukan pada indeks {indeks}.")
else:
  print(f"Huruf '{target}' tidak ditemukan dalam deret huruf.")
```

```python

def interpolation_search(list_, target):
  """
  Fungsi untuk mencari elemen dalam daftar yang telah diurutkan menggunakan metode interpolation search.

  Args:
    list_: Daftar yang telah diurutkan yang akan dicari.
    target: Elemen yang akan dicari.

  Returns:
    Indeks elemen jika ditemukan, -1 jika tidak ditemukan.
  """
  low = 0
  high = len(list_) - 1

  while low <= high and target >= list_[low] and target <= list_[high]:
    pos = low + int(((float(high - low) / (list_[high] - list_[low])) * (target - list_[low])))

    if list_[pos] == target:
      return pos

    if list_[pos] < target:
      low = pos + 1
    else:
      high = pos - 1

  return -1

deret_huruf = ['y', 'u', 'i', 'w', 'o', 'a', 'q', 'u', 'j', 'p']
deret_huruf.sort()
target = 'u'

indeks = interpolation_search(deret_huruf, target)

if indeks != -1:
  print(f"Huruf '{target}' ditemukan pada indeks {indeks}.")
else:
  print(f"Huruf '{target}' tidak ditemukan dalam deret huruf.")
```
