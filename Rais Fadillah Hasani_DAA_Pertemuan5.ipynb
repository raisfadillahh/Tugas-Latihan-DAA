```python
# A Divide and Conquer based program for maximum subarray sum problem
# Find the maximum possible sum in arr[] such that arr[m] is part of it
def maxCrossingSum(arr, low, mid, high):
    # Include elements on left of mid
    left_sum = float('-inf')
    sm = 0

    for i in range(mid, low - 1, -1):
        sm += arr[i]
        if sm > left_sum:
            left_sum = sm

    # Include elements on right of mid
    right_sum = float('-inf')
    sm = 0

    for i in range(mid + 1, high + 1):
        sm += arr[i]
        if sm > right_sum:
            right_sum = sm

    # Return sum of elements on left and right of mid
    # Returning only left_sum + right_sum will fail for [-2, 1]
    return max(left_sum + right_sum, left_sum, right_sum)

# Example usage
arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
low = 0
high = len(arr) - 1
mid = (low + high) // 2  # Example midpoint

result = maxCrossingSum(arr, low, mid, high)
print("Maximum crossing sum:", result)
```

```python
def maxCrossingSum(arr, l, m, h):
    # Mencari jumlah maksimum subarray yang melintasi titik tengah
    left_sum = float('-inf')
    right_sum = float('-inf')

    # Mencari maksimum dari subarray di sebelah kiri m
    current_sum = 0
    for i in range(m, l - 1, -1):
        current_sum += arr[i]
        if current_sum > left_sum:
            left_sum = current_sum

    # Mencari maksimum dari subarray di sebelah kanan m
    current_sum = 0
    for i in range(m + 1, h + 1):
        current_sum += arr[i]
        if current_sum > right_sum:
            right_sum = current_sum

    # Mengembalikan jumlah maksimum yang melintasi m
    return left_sum + right_sum

def maxSubArraySum(arr, l, h):
    # Invalid Range: low is greater than high
    if l > h:
        return float('-inf')

    # Base Case: Only one element
    if l == h:
        return arr[l]

    # Find middle point
    m = (l + h) // 2

    # Return maximum of three possible cases
    return max(maxSubArraySum(arr, l, m),
               maxSubArraySum(arr, m + 1, h),
               maxCrossingSum(arr, l, m, h))

# Driver Code
arr1 = [4, 12, 23, 9, 21, 1, 35, 2, 24]
n1 = len(arr1)
max_sum1 = maxSubArraySum(arr1, 0, n1 - 1)
print("Maximum contiguous sum is", max_sum1)

arr2 = [12, 51, 23, 84, 5, 9]
n2 = len(arr2)
max_sum2 = maxSubArraySum(arr2, 0, n2 - 1)
print("Maximum contiguous sum is", max_sum2)
```

```python
def divideAndConquer_Max(arr, ind, length):
    # Menghitung nilai maksimum dalam array
    if ind >= length - 2:
        if arr[ind] > arr[ind + 1]:
            return arr[ind]
        else:
            return arr[ind + 1]

    # Rekursi untuk bagian selanjutnya
    maximum = divideAndConquer_Max(arr, ind + 1, length)

    if arr[ind] > maximum:
        return arr[ind]
    else:
        return maximum

def divideAndConquer_Min(arr, ind, length):
    # Menghitung nilai minimum dalam array
    if ind >= length - 2:
        if arr[ind] < arr[ind + 1]:
            return arr[ind]
        else:
            return arr[ind + 1]

    # Rekursi untuk bagian selanjutnya
    minimum = divideAndConquer_Min(arr, ind + 1, length)

    if arr[ind] < minimum:
        return arr[ind]
    else:
        return minimum

# Inisialisasi array
arr = [6, 4, 8, 90, 12, 56, 7, 1, 63]

# Mendapatkan nilai maksimum dan minimum
maximum = divideAndConquer_Max(arr, 0, len(arr))
minimum = divideAndConquer_Min(arr, 0, len(arr))

# Mencetak hasil
print("The minimum number in the array is:", minimum)
print("The maximum number in the array is:", maximum)
```

```python
def mergeSort(array):
    # Jika panjang array lebih dari 1, lakukan pengurutan
    if len(array) > 1:
        # Titik pemisah untuk membagi array menjadi dua subarray
        r = len(array) // 2
        L = array[:r]  # Subarray kiri
        M = array[r:]  # Subarray kanan

        # Urutkan kedua bagian
        mergeSort(L)
        mergeSort(M)

        i = j = k = 0

        # Sampai kita mencapai salah satu akhir dari L atau M,
        # pilih yang lebih kecil antara elemen L dan M dan letakkan di posisi yang benar di array
        while i < len(L) and j < len(M):
            if L[i] < M[j]:
                array[k] = L[i]
                i += 1
            else:
                array[k] = M[j]
                j += 1
            k += 1

        # Ketika kita kehabisan elemen di salah satu dari L atau M,
        # ambil elemen yang tersisa dan letakkan di array
        while i < len(L):
            array[k] = L[i]
            i += 1
            k += 1

        while j < len(M):
            array[k] = M[j]
            j += 1
            k += 1

# Fungsi untuk mencetak array
def printList(array):
    for i in range(len(array)):
        print(array[i], end=" ")
    print()

# Inisialisasi array
array = [8, 3, 2, 9, 7, 1, 5, 4]
mergeSort(array)  # Memanggil fungsi mergeSort
print("Sorted array is: ")
printList(array)  # Mencetak array yang sudah terurut
```

```python
# Insertion Sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        # Pindahkan elemen yang lebih besar dari key ke satu posisi di depan dari posisi saat ini
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return arr

# Contoh penggunaan
arr = [4, 12, 23, 9, 21, 1, 5, 2]
print("Sorted array is:", insertion_sort(arr))
```

```python
# Function to find the partition position
def partition(array, low, high):
    # Choose the rightmost element as pivot
    pivot = array[high]
    # Pointer for greater element
    i = low - 1

    # Traverse through all elements
    for j in range(low, high):
        # Compare each element with pivot
        if array[j] <= pivot:
            # If element smaller than pivot is found
            # Swap it with the greater element pointed by i
            i += 1
            # Swapping element at i with element at j
            array[i], array[j] = array[j], array[i]

    # Swap the pivot element with the greater element specified by i
    array[i + 1], array[high] = array[high], array[i + 1]

    # Return the position from where partition is done
    return i + 1

# Function to perform quicksort
def quickSort(array, low, high):
    if low < high:
        # Find pivot element such that
        # Elements smaller than pivot are on the left
        # Elements greater than pivot are on the right
        pi = partition(array, low, high)

        # Recursive call on the left of pivot
        quickSort(array, low, pi - 1)
        # Recursive call on the right of pivot
        quickSort(array, pi + 1, high)

# Example usage
data = [1, 7, 4, 1, 10, 9, -2]
print("Unsorted Array:")
print(data)

size = len(data)
quickSort(data, 0, size - 1)

print('Sorted Array in Ascending Order:')
print(data)
```

```python
# Fungsi untuk menghitung jumlah inversi
def countInversion(arr):
    result = 0
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] > arr[j]:
                result += 1
    return result

# Contoh penggunaan
arr = [21, 70, 36, 14, 25]
result = countInversion(arr)
print("Jumlah inversi adalah:", result)
```

```python
# Hitung Inversi dengan metode divide dan conquer
def countInversion(arr):
    # Inisialisasi jumlah inversi
    icount = 0

    # Basis: jika panjang array kurang dari atau sama dengan 1
    if len(arr) <= 1:
        return icount

    # Temukan titik tengah
    mid = len(arr) // 2

    # Bagi array menjadi dua bagian
    left = arr[:mid]
    right = arr[mid:]

    # Hitung inversi di kedua bagian
    icount += countInversion(left)
    icount += countInversion(right)

    # Indeks untuk menggabungkan
    i = j = k = 0

    # Gabungkan kedua bagian
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            arr[k] = left[i]
            i += 1
        else:
            # Jika elemen dari kanan lebih kecil, itu menunjukkan inversi
            arr[k] = right[j]
            icount += (mid - i)  # Semua elemen yang tersisa di 'left' adalah inversi
            j += 1
        k += 1

    # Ambil sisa elemen dari left
    while i < len(left):
        arr[k] = left[i]
        i += 1
        k += 1

    # Ambil sisa elemen dari right
    while j < len(right):
        arr[k] = right[j]
        j += 1
        k += 1

    return icount

# Contoh penggunaan
arr = [1, 20, 6, 4, 5]
result = countInversion(arr)
print("Jumlah inversi adalah:", result)
```

```python
# Fungsi untuk mencari jumlah maksimum subarray
def maxSubSum(arr):
    max_so_far = 0
    max_ending_here = 0

    for i in range(len(arr)):
        max_ending_here += arr[i]

        # Memperbarui max_so_far jika max_ending_here lebih besar
        if max_ending_here > max_so_far:
            max_so_far = max_ending_here

        # Reset max_ending_here jika nilainya negatif
        if max_ending_here < 0:
            max_ending_here = 0

    return max_so_far

# Contoh penggunaan
arr = [-2, -5, 6, -2, -3, 1, 5, -6]
result = maxSubSum(arr)
print("Jumlah maksimum subarray adalah:", result)
```

```python
def longestCommonPrefix(arr):
    size = len(arr)

    # Jika ukuran array 0, kembalikan string kosong
    if size == 0:
        return ""

    # Jika hanya ada satu string, kembalikan string tersebut
    if size == 1:
        return arr[0]

    # Urutkan array string
    arr.sort()

    # Temukan panjang minimum dari string pertama dan terakhir setelah pengurutan
    end = min(len(arr[0]), len(arr[size - 1]))

    # Temukan prefix umum antara string pertama dan terakhir
    i = 0
    while i < end and arr[0][i] == arr[size - 1][i]:
        i += 1

    # Kembalikan prefix umum
    return arr[0][0:i]

# Contoh penggunaan
arr1 = ["geeksforgeeks", "geeks", "geek", "geezer"]
result1 = longestCommonPrefix(arr1)
print("", result1)  # Output: "gee"
```

```python
def medianOfArray(arr1, arr2, n):
    m1 = -1  # Nilai median sebelumnya
    m2 = -1  # Nilai median saat ini
    count = 0
    i = j = 0

    while count < n + 1:
        count += 1

        # Jika semua elemen dari arr1 telah diiterasi
        if i == n:
            m1 = m2
            m2 = arr2[0]
            break

        # Jika semua elemen dari arr2 telah diiterasi
        if j == n:
            m1 = m2
            m2 = arr1[0]
            break

        # Bandingkan elemen dari kedua array
        if arr1[i] < arr2[j]:
            m1 = m2  # Simpan nilai median sebelumnya
            m2 = arr1[i]  # Ambil nilai median saat ini
            i += 1
        else:
            m1 = m2  # Simpan nilai median sebelumnya
            m2 = arr2[j]  # Ambil nilai median saat ini
            j += 1

    # Mengembalikan median dari dua array
    return (m1 + m2) / 2  # Menggunakan pembagian float untuk median yang tepat

# Contoh penggunaan
arr1 = [1, 12, 15, 26, 38]
arr2 = [2, 13, 17, 30, 45]
print("Median dari dua array adalah:", medianOfArray(arr1, arr2, len(arr1)))  # Output: 16
```

```python
def Solution(arr):
    n = len(arr)

    # Jika panjang array genap
    if n % 2 == 0:
        z = n // 2
        e = arr[z]     # Elemen tengah
        q = arr[z - 1] # Elemen tengah sebelumnya
        ans = (e + q) / 2
        return ans
    # Jika panjang array ganjil
    else:
        z = n // 2
        ans = arr[z]   # Elemen tengah
        return ans

# Contoh penggunaan
arr1 = [-5, 3, 6, 12, 15]
arr2 = [-12, -10, -6, -3, 4, 10]

# Menggabungkan kedua array
arr3 = arr1 + arr2

# Mengurutkan array hasil gabungan
arr3.sort()

# Menampilkan median
print("Median =", Solution(arr3))  # Output: Median = 3
```

```python
def floorSorted(arr, low, high, x):
    # Jika low lebih besar dari high, x tidak ditemukan
    if low > high:
        return -1

    # Jika elemen terkecil lebih besar dari x, tidak ada floor
    if arr[low] > x:
        return -1

    # Jika elemen terbesar kurang dari atau sama dengan x, kembalikan elemen terbesar
    if arr[high] <= x:
        return arr[high]

    mid = (low + high) // 2

    # Jika elemen tengah adalah x, maka floor adalah x itu sendiri
    if arr[mid] == x:
        return arr[mid]

    # Jika mid lebih besar dari 0 dan x lebih besar dari elemen sebelumnya
    if mid > 0 and x >= arr[mid - 1] and arr[mid] > x:
        return arr[mid - 1]

    # Jika mid kurang dari high dan x kurang dari elemen berikutnya
    if mid < high and x < arr[mid + 1] and x >= arr[mid]:
        return arr[mid]

    # Jika x lebih besar dari elemen tengah
    if x > arr[mid]:
        return floorSorted(arr, mid + 1, high, x)

    # Jika x kurang dari elemen tengah
    return floorSorted(arr, low, mid - 1, x)

# Contoh penggunaan
arr = [1, 2, 8, 10, 12, 14, 19]
x = 5
result = floorSorted(arr, 0, len(arr) - 1, x)
print("Floor dari", x, "adalah:", result)  # Output: 2
```

```python
def closestNumber(arr, low, high, x):
    # Jika rentang tidak valid
    if low > high:
        return -1

    # Jika semua elemen kurang dari atau sama dengan x
    if arr[high] <= x:
        return arr[high]

    # Jika semua elemen lebih besar dari atau sama dengan x
    if arr[low] >= x:
        return arr[low]

    mid = (low + high) // 2

    # Jika elemen tengah adalah x
    if arr[mid] == x:
        return arr[mid]

    # Mencari nilai absolut dari selisih
    abs_mid = abs(arr[mid] - x)

    # Mencari nilai terdekat di sebelah kiri
    if mid > 0:
        abs_left = abs(arr[mid - 1] - x)
        if abs_left < abs_mid:
            return closestNumber(arr, low, mid - 1, x)

    # Mencari nilai terdekat di sebelah kanan
    if mid < high:
        abs_right = abs(arr[mid + 1] - x)
        if abs_right < abs_mid:
            return closestNumber(arr, mid + 1, high, x)

    return arr[mid]

# Contoh penggunaan
arr = [2, 5, 6, 7, 8, 8, 9]
x = 9
result = closestNumber(arr, 0, len(arr) - 1, x)
print("Nilai terdekat dengan", x, "adalah:", result)  # Output: 9
```

```python
def find_closest(lst, k):
    # Mengurutkan list
    lst.sort()
    closest_num = lst[0]

    for num in lst:
        # Jika selisih dari num dan k lebih kecil dari closest_num dan k
        if abs(num - k) < abs(closest_num - k):
            closest_num = num
        # Jika num sudah lebih besar dari k, kita sudah menemukan yang terdekat
        if num > k:
            break

    return closest_num

# Contoh penggunaan
lst = [3.64, 5.2, 9.42, 9.35, 8.5, 8]
k = 9.1
print(find_closest(lst, k))  # Output: 9.35

lst = [2, 5, 5, 7, 8, 8, 9]
k = 6
print(find_closest(lst, k))  # Output: 5
```

```python
def fixedPoint(arr, low, high):
    # Jika rentang tidak valid
    if low > high:
        return -1

    # Memeriksa jika elemen di indeks high adalah fixed point
    if arr[high] == high:
        return arr[high]

    # Memeriksa jika elemen di indeks low adalah fixed point
    if arr[low] == low:
        return arr[low]

    mid = (low + high) // 2

    # Jika elemen tengah adalah fixed point
    if arr[mid] == mid:
        return arr[mid]

    # Jika nilai di mid lebih besar dari mid, cari di sebelah kiri
    if mid > arr[mid]:
        return fixedPoint(arr, low, mid - 1)

    # Jika nilai di mid kurang dari mid, cari di sebelah kanan
    return fixedPoint(arr, mid + 1, high)

# Contoh penggunaan
arr = [9, 1, 4, 5, 2]
result = fixedPoint(arr, 0, len(arr) - 1)
print("Fixed point:", result)  # Output: 1
```

```python
def bagi_bilangan(arr):
    # Menentukan pembagian manual untuk memastikan 21 di grup kiri
    grup1 = arr[:5]  # Mengambil 5 elemen pertama
    grup2 = arr[5:]  # Mengambil sisa elemen

    # Mengonversi bagian menjadi string untuk ditampilkan
    grup1_str = '+'.join(map(str, grup1))
    grup2_str = '+'.join(map(str, grup2))

    # Mengembalikan hasil
    return f"({grup1_str}) + ({grup2_str})"

# Daftar bilangan
bilangan = [4, 12, 23, 9, 21, 1, 35, 2, 24]

# Memanggil fungsi dan mencetak hasil
hasil = bagi_bilangan(bilangan)
print("Pembagian bilangan:", hasil)
```
